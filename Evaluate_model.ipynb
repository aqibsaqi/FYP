```python
import gradio as gr
import tensorflow as tf
import numpy as np

# Image dimensions expected by the model
img_height = 256  # Adjusted for MobileNetV2 input
img_width = 256   # Adjusted for MobileNetV2 input

#  class names
class_names = ['Cyst', 'Normal', 'Stone', 'Tumor']
# Prediction function
def classify_image(img):
    # Resize image to 256x256
    img = tf.image.resize(img, [img_height, img_width])

    # Convert grayscale images to RGB
    if img.shape[-1] == 1:  # Check the image is grayscale
        img = tf.image.grayscale_to_rgb(img)

    # Normalize the image to [0, 1]
    img = img / 255.0

    # Adjust the contrast (optional)
    img = tf.image.adjust_contrast(img, 2.0)

    # Add batch dimension to match the model's expected input shape
    img = tf.expand_dims(img, 0)

    # Make the prediction
    pred = model.predict(img)

    # Get the class with the highest probability
    predicted_class = np.argmax(pred)

    # Return the class name and its probability
    return {class_names[predicted_class]: float(pred[0][predicted_class])}

# Gradio interface
demo = gr.Interface(
    fn=classify_image,  # Prediction function
    inputs=gr.Image(type="numpy"),  # Image input
    outputs=gr.Label(num_top_classes=4),  # Output the top 4 predicted classes
    title="Kidney Stone Classifier",  # Title for the interface
    description="Upload a CT kidney image to classify it as Normal, Cyst, Tumor, or Stone."  # Description for the user
)

# Launch the interface
demo.launch(share=True, debug=True)
```
